```python
import pandas as pd
import re
import string

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Biocreative/Biocreative/litcovid_dataset.csv')
# df = df.dropna()
df.head()
```

```python
def clean_text(text):

  "Make text lower, remove text in square brackets, remove links, remove punctutations and remove words containing numbers"

  text = text.lower()
  text = re.sub("\[.*?\]", '', text)
  text = re.sub("https?://\S+|www\.\S+", '', text)
  text = re.sub("<.*?>+", '', text)
  text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
  text = re.sub("\n", "", text)
  text = re.sub("\w*\d\w", "", text)
  return text
```

```python
abstract = df["abstract"][0]
# print(abstract)
clean_text(abstract)
```

```python
abstract
```

```python
def text_preprocessing(text):

  "Cleaning and parsing the text"

  tokenizer = nltk.tokenize.RegexpTokenizer(r"\w+")
  nopunc = clean_text(text)
  tokenized_text = tokenizer.tokenize(nopunc)
  # remove stopwords
  processed_text = [w for w in tokenized_text if w not in stopwords.words('english')]
  combined_text = ' '.join(processed_text)
  return combined_text
```

```python
text_preprocessing(abstract)
```

```python
df2 = df.abstract
df3 = df2.apply(text_preprocessing) 
```

```python
df4 = df[df.columns[9:]]
df4
```

```python
# concat dataframes
frame = [df3, df4]
result = pd.concat(frame, axis=1)
```
